```python
#Date 11/01/22
import pandas as pd
import numpy as np

df_sales=pd.read_excel('C:/Users/przir/OneDrive/Desktop/Data Science/Libraries/DataSets/SaleData.xlsx')
df_sales.head()
```

```python
#1-create a Pivot table with multiple indexes from a given excel sheet
pd.pivot_table(df_sales,index=['Region','SalesMan'])
```

```python
#2-create a Pivot table and find the total sale amount region wise, manager wise
print(pd.pivot_table(df_sales,index=['Region','SalesMan'],values=['Sale_amt'],aggfunc=np.sum))
```

```python
#3-create a Pivot table and find the total sale amount region wise, manager wise, sales man wise
print(pd.pivot_table(df_sales,index=['Region','Manager','SalesMan'],values=['Sale_amt'],aggfunc=np.sum))
```

```python
#4-create a Pivot table and find the item wise unit sold.
#print(pd.pivot_table(df_sales,index=['Item','Units']))
#print(pd.pivot_table(df_sales,index=['Item'],values='Units'))
print(pd.pivot_table(df_sales,index=['Item'],values='Units',aggfunc=np.sum))
```

```python
#5- create a Pivot table and find the region wise total sale.
print(pd.pivot_table(df_sales,index=['Region'],values='Sale_amt',aggfunc=np.sum))
```

```python
#6-Pivot table and find the region wise, item wise unit sold
print(pd.pivot_table(df_sales,index=['Region','Item'],values='Units',aggfunc=np.sum))
```

```python
#7-create a Pivot table and count the manager wise sale and mean value of sale amount.
print(pd.pivot_table(df_sales,index=['Manager'],values='Sale_amt',aggfunc=[np.mean,len]))
```

```python
#8-create a Pivot table and find manager wise, salesman wise total sale and also display the sum of all sale amount at the bottom
print(pd.pivot_table(df_sales,index=['Manager','SalesMan'],values=['Units','Sale_amt'],aggfunc=np.sum,margins=True,fill_value=0))
```

```python
#9-create a Pivot table and find the total sale amount region wise, manager wise, sales man wise where Manager = "Douglas"
table=pd.pivot_table(df_sales,index=['Region','Manager','SalesMan'],values='Sale_amt',aggfunc=np.sum)
print(table.query('Manager==["Douglas"]'))
```

```python
#10- create a Pivot table and find the region wise Television and Home Theater sold.
table=pd.pivot_table(df_sales,index=['Region','Item'],values='Units')  #values always integers
print(table.query('Item==["Home Theater","Television"]'))
```

```python
#11-create a Pivot table and find the maximum sale value of the items
print(pd.pivot_table(df_sales,index=['Item'],values='Sale_amt',aggfunc=np.max))
```

```python
#12-create a Pivot table and find the minimum sale value of the items
print(pd.pivot_table(df_sales,index=['Item'],values='Sale_amt',aggfunc=np.min))
```

```python
#13-create a Pivot table and find the maximum and minimum sale value of the items.
table=pd.pivot_table(df_sales,index=['Item'],values='Sale_amt',aggfunc=[np.max,np.min])
table
```

```python
import pandas as pd
import numpy as np

df_titanic=pd.read_csv('C:/Users/przir/OneDrive/Desktop/Data Science/Libraries/DataSets/titanic.csv')
print(df_titanic.head())
```

```python
#1- print a concise summary of the dataset 
df_titanic.info()
```

```python
#2-extract the column labels, shape and data types
print('\ncolumns summary')
print(df_titanic.columns)
print('\nshape of dateset')
print(df_titanic.shape)
print('\ndata type')
print(df_titanic.dtypes)
```

```python
#3-create a Pivot table with multiple indexes
result=pd.pivot_table(df_titanic,index=['survived','sex'],aggfunc=np.sum)
result
```

```python
#4-Pivot table and find survival rate by gender on various classes
#result=pd.pivot_tale(df_titanic,index=['survived','sex'],columns='class',aggfunc=np.sum)
result=df_titanic.pivot_table('survived',index='sex',columns='class')
result
```

```python
#5-create a Pivot table and find survival rate by gender
result=df_titanic.groupby('sex')[['survived']].mean()
result
```

```python
#6-create a Pivot table and find survival rate by gender, age wise of various classes.
result=df_titanic.pivot_table('survived',index=['sex','age'],columns='class')
result
```

```python
#7-partition each of the passengers into four categories based on their age
result=pd.cut(df_titanic['age'],[0,10,30,60,80])
result
```

```python
#8-create a Pivot table and count survival by gender, categories wise age of various classes
age=pd.cut(df_titanic['age'],[0, 10, 30, 60, 80])

result=df_titanic.pivot_table('survived',index=['sex',age],columns='pclass',aggfunc='count')
result
```

```python
#9-create a Pivot table and find survival rate by gender, age of the different categories of various classes
age=pd.cut(df_titanic['age'],[0,20,55])
result=df_titanic.pivot_table('survived',index=['sex',age],columns='class')
result
```

```python
#Date-11/2/22
#10-create a Pivot table and find survival rate by gender, 
#age of the different categories of various classes. Add the fare as a dimension of columns
#and partition fare column into 2 categories based on the values present in fare columns.
import pandas as pd
import numpy as np
fare=pd.qcut(df_titanic['fare'],2)
age=pd.cut(df_titanic['age'],[0, 10, 30, 60, 80])
result=df_titanic.pivot_table('survived',index=['sex',age],columns=[fare,'pclass'])
print(result)
```

```python
#11- create a Pivot table and calculate number of women and men were in a particular cabin class. 
import pandas as pd
import numpy as np
result=df_titanic.pivot_table(index=['sex'],columns=['pclass'],aggfunc='count')
result
```

```python
#12- create a Pivot table and find survival of both gender and class affected
import pandas as pd
import numpy as np

result=df_titanic.groupby(['sex','class'])['survived'].aggregate('mean').unstack()
result
```

```python
#13- create a Pivot table and compute survival totals of all classes along each group
result=df_titanic.pivot_table('survived',index='sex',columns='class',margins=True)
result
```

```python
#14-create a Pivot table and calculate how many women and men were in a particular cabin class
result=df_titanic.pivot_table(index='sex',columns=['pclass'],values='survived',aggfunc='count')
result
```

```python
#15- create a Pivot table and find number of survivors and average rate grouped by gender and class.
result=df_titanic.pivot_table(index='sex',columns='class',aggfunc={'survived':sum,'fare':'mean'})
result
```

```python
#16- create a Pivot table and find number of adult male, adult female and children
result=pd.pivot_table(df_titanic,index=['sex','who'],aggfunc='count')
result
```

```python
result = df_titanic.pivot_table('sex',  'who', aggfunc = 'count')
print(result)
```

```python
#17-create a Pivot table and check missing values of children
result=df_titanic[df_titanic['who']=='child'].isnull().sum()
result
```

```python
#18-create a Pivot table and separate the gender according to whether 
#they traveled alone or not to get the probability of survival
result=df_titanic.pivot_table('survived',index=['sex','alone'],columns='class')
result
```

```python
#19-create a Pivot table and find the probability of survival by class, gender, solo boarding and port of embarkation
result=df_titanic.pivot_table('survived',['sex','alone'],['embark_town', 'class'])
result
```
