```python
#22-import three datasheets from a given excel data (coalpublic2013.xlsx ) and combine in to a single dataframe
import pandas as pd

df1=pd.read_excel('C:/Users/przir/OneDrive/Desktop/Data Science/Libraries/DataSets/employee.xlsx',sheet_name=1)
df2=pd.read_excel('C:/Users/przir/OneDrive/Desktop/Data Science/Libraries/DataSets/employee.xlsx',sheet_name=2)
df3=pd.read_excel('C:/Users/przir/OneDrive/Desktop/Data Science/Libraries/DataSets/employee.xlsx',sheet_name=0)

df=pd.concat([df1,df2,df3])
df.reset_index(drop=True)
```

```python
#23-draw a bar plot where each bar will represent one of the top 10 production
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

sorted_by_production=df.sort_values(['Labor Hours'],ascending=False).head(10)
sorted_by_production['Labor Hours'].head(10).plot(kind='barh')
plt.show()
```
